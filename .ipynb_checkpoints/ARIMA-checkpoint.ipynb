```python
# 导入相关库
import torch
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
```

```python
# 相关系数图
x = np.loadtxt('clipboard')
correlation_matrix = np.corrcoef(x, rowvar=False)
fig, ax = plt.subplots(figsize=(6, 6))
cax = ax.matshow(correlation_matrix, cmap='coolwarm')
plt.colorbar(cax)
plt.show()
```

```python

# 随机森林特征重要性排序
wine = np.loadtxt('clipboard')
np.random.shuffle(wine)
train_sub = int(0.7 * len(wine))
train_data = wine[:train_sub]
test_data = wine[train_sub:]
X_train, y_train = train_data[:, 1:], train_data[:, 0]
rf = RandomForestRegressor(n_estimators=500, max_features=3)
rf.fit(X_train, y_train)
importance = rf.feature_importances_
plt.bar(range(1, importance.size+1), importance)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Variable Importance')
plt.show()
```

```python
# 神经网络
x = np.loadtxt('clipboard')
np.random.shuffle(x)
train_data = x[:146, :]
test_data = x[146:, :]
X_train, y_train = train_data[:, 1:], train_data[:, 0]
X_test, y_test = test_data[:, 1:], test_data[:, 0]
model = MLPRegressor(hidden_layer_sizes=(5, 4), activation='tanh', solver='lbfgs')
model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
y_train_pred = y_train_pred.reshape(-1, 1)
y_test_pred = y_test_pred.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
train_rmse = np.sqrt(mean_squared_error(y_train_pred, y_train))
test_rmse = np.sqrt(mean_squared_error(y_test_pred, y_test))
print('训练集RMSE:', train_rmse)
print('测试集RMSE:', test_rmse)
```
